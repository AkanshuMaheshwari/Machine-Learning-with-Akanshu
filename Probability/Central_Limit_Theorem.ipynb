## Central Limit Theorem

##### 6. Simulate the Central Limit Theorem by following these steps  
  a. Generate 10,000 random numbers from a uniform distribution.  
  b. Draw 1000 samples of size n = 30.  
  c. Calculate and visualize the distribution of sample means.  
  Steps  
      a. Use numpy.random.uniform().  
      b. Plot both the uniform distribution and the sample mean distribution for comparison.


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def simulate_clt(size = 30, num_sample = 1000, random_size = 10000):

    # generate 10000 random numbers from a uniformm distribution
    population = np.random.uniform(low = 0, high = 10, size = random_size)

    # draw 1000 samples of size n =30 & calculate sample mean
    sample_means = [np.mean(np.random.choice(population, size = size)) for _ in range(num_sample)]

    plt.figure(figsize=(12,5))   # visualize distribution

    # plot uniform distribution
    plt.subplot(1,2,1)
    sns.histplot(population, bins= 30,  color= 'blue', alpha = 0.7, edgecolor = 'black', stat = 'density')
    plt.title("Uniform Distribution")
    plt.xlabel("Value")
    plt.ylabel("Density")

    # Plot sample mean distribution
    plt.subplot(1, 2, 2)
    sns.histplot(sample_means, bins=30, color="red", alpha=0.7, edgecolor="black", stat="density")
    plt.title("Distribution of Sample Means (CLT Effect)")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")

    plt.tight_layout()
    plt.show()

# Run the simulation
simulate_clt()
```
