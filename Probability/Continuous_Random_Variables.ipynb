## Continuous Random Variables

##### 5. Simulate 2000 random samples from an exponential distribution with a mean of 5. Visualize the distribution using:  
  a. A histogram.  
  b. A probability density function (PDF) overlay.  
  Steps  
      a. Use numpy.random.exponential().  
      b. Use matplotlib to create visualizations.  


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import expon
```

```python
def simulate_exponential_distribution(size = 2000, mean = 5):

    # generate random sample
    samples = np.random.exponential(scale = mean, size = size)

    # create histogram
    sns.histplot(samples, bins = 30, kde = False, color= 'lightblue', alpha = 0.7, edgecolor = ('black'), stat = 'density')

    # overlay probability density function (PDF)
    x = np.linspace(0, np.max(samples), 100)
    pdf_values = expon.pdf(x, scale = mean)
    plt.plot(x, pdf_values, color = 'red', label = 'Exponential PDF')

    # add labels and title
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.title('Histogram & PDF of exponential distribution')
    plt.legend()
    plt.grid()

    # show plot
    plt.show()

# run the function
simulate_exponential_distribution()
```
